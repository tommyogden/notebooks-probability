# Event Spaces, Probability and Independent Events

## Event Spaces

We define events by an agreed _event space_ $\Omega$, e.g. single die throw: $\Omega = \{ 1,2,3,4,5,6 \}$. 

We also define set of _measurable events_ $S$ to assign probabilities. Each _event_ $A \in S$ is a subset of $\Omega$, e.g. $A = \{ 1,3,5 \}$ is the event representing 'throw an odd' or $A = \{ 6 \}$ representing 'throw a six'.

### Definition

Here are the properties of an **event space**:

- Contains _empty event_ $\emptyset$ and the _trivial event_ $\Omega$
- _Closed under union_: if $A,B \in S$ then $A \cup B \in S$
- _Closed under complementation_: if $A \in S$ then $\Omega - A \in S$

These also imply _closed under intersection_ and _closed under set difference_.

The basic idea is that it's a measure space such that the measure of the whole space is unity.

## Probability Distributions

### Definition

A **probability distribution** $P$ over $(\Omega, S)$ is a mapping from events in $S$ to real values such that

$$
\begin{align}
& P(A) \ge 0, \forall A \in S \tag{1a} \\
& P(\Omega) = 1 \tag{1b}
\end{align}
$$

and if $A,B \in S$ and $A \cap B \ne \emptyset$ then 

$$P(A \cup B) = P(A) + P(B) \tag{1c}.$$

Other conditions including $P(\emptyset) = 0$ and $P(A \cup B) = P(A) + P(B) - P(A \cap B)$ can be derived from these.

### Corollary

$$
P(\emptyset) = 0 \tag{2}
$$

#### Proof

$$
\begin{align}
& \Omega \cup \emptyset = \Omega \\
\implies & P(\Omega \cup \emptyset) = P(\Omega) + P(\emptyset) \\
\implies & P(\Omega) = P(\Omega) + P(\emptyset) \\
\implies & P(\emptyset) = 0. \tag*{$\blacksquare$}
\end{align}
$$

### Corollary

If $A,B \in S$ and $A \subseteq B$ then $P(A) \le P(B)$.    

    +------------------------+
    |                       S|
    | +--------------+       |
    | |B             |       |
    | |      +-----+ |       |
    | |      |A    | |       |
    | |      +-----+ |       |
    | +--------------+       |
    |                        |
    +------------------------+

#### Proof

Let $C = B - (B \cap A) = B - A$.  

Then $A \cap C = 0 \implies P(B) = P(A) + P(C)$ and as we have $P(C) \ge 0$,  

$$
\implies P(A) \le P(B). \tag*{$\blacksquare$}
$$  


### Corollary

$$
P(A \cup B) = P(A) + P(B) - P(A \cap B) \tag{3}
$$

    +------------------------+
    |                       S|
    | +--------------+       |
    | |B             |       |
    | |       +-----------+  |
    | |       |      |   A|  |
    | +--------------+    |  |
    |         +-----------+  |
    |                        |
    +------------------------+

#### Proof

We'll write $A \cap B = AB$ for short.

Note that $A \cup B = (A B^c) \cup (A B) \cup (A^c B)$ and these three events are disjoint. Thus by $(3)$ we find

$$
\begin{align}
P(A \cup B) & = P(A \cup B = (A B^c) \cup (A B) \cup (A^c B)) \\
& = P(A B^c) + P(A B) + P (A^c B) \\
& = P(A B^c) + P(A B) + P (A^c B) + P(AB) - P(AB) \\
& = P( (A B^c) \cup (A B)) + P( (A B) \cup (A^c B)) - P(AB) \\
& = P(A) + P(B) - P(AB). \tag*{$\blacksquare$}
\end{align}
$$

## Independent Events

### Definition

Two events $A$ and $B$ are **independent** if $P(AB) = P(A)P(B)$.

Sometimes we start by assuming that two events are independent (e.g. throwing a six on subsequent dice throws). Sometimes we use the above definition to determine independence.

Note that two disjoint non-zero-probability events cannot be independent as $P(AB) = 0$ but we must have $P(A)P(B) > 0$.
